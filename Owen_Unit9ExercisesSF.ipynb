Explain what happens when we do matrix multiplication.

Yes, I know I basically already asked you to do this in the notes. Don't copy and paste your anwser from before; honestly do the exercise again.

We takes the columns of the second matrix and line them up with the rows of the first, and then shift the matrices across eachother, multiplying corresponding entries and adding them up to get the entries in our final result matrix. The dimensions of the output are determined by the number of rows in the first matrix and the number of columns in the second. Matrices can only be multiplied when the number of columns in the first matrix is the same as the number of rows in the second.

Explain what part of the neural network visualization is matrix multiplication.

Matrix multiplication is all of the links between different nodes that travel between layers. Each of these links has an associated weight, which is the matrix that you multiply the data by to reach the next node (similar to how our linear model had one weight matrix to go from x, our predictor, to y, our predicted, except there's a lot more than just one weight matrix here).
